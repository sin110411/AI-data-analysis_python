```python
#행선택 iloc와 loc

df=pd.DataFrame({"A":[1,2,3],"B":[4,5,6],"C":[7,8,9]},index=['a','b','c'])
print(df)
print()

#1) loc를 사용한 행선택 : loc(행이름)
print(df.loc['a'],type(df.loc['a'])) #행선택  #타입 시리즈
print()

#1-1)iloc를 사용한 행선택
print(df.iloc[0]) #결과같음 #타입 시리즈
print()

#2)여러행 선택
print(df.loc[['a','c']],type(df.loc[['a','c']])) #여러행 선택이기떄문에 멀티인덱스 사용
#데이터 프레임으로 반환
print(df.loc['a':'b']) #슬라이싱 #이름으로 가져오면 끝값 포함
print()

#2-1) loc여러행선택
print(df.iloc[[0,2]]) #데이터 프레임 형태
print(df.iloc[:2]) #번호 슬라이싱이라 끝값 포함 안됨
print()

#iloc와 loc의 차이점
#1. iloc는 정수인덱스(번호기반)으로 선택하고 loc는 인덱스 이름을 기반으로 선택
#2. iloc는 범위끝포함X loc는 범위끝 포함
#3. loc메소드를 사용할때는 인덱스 이름값을 지정행해야함!!
```

```python
#원소 선택하기

stu_data = {'이름' : ['짱구', "철수", "유리"],
           'python' : [60, 90, 80],
            'java' : [70, 100, 85],
            'C언어' : [50, 85, 70],
            '웹표준' : [80, 95, 90] 
           }
df=pd.DataFrame(stu_data)
print(df)
print()

#특정열을 행으로 지정하기 : set_index(열이름)
df.set_index('이름',inplace=True)
print(df)
print()
#iloc[행번호, 열번호]
#loc[행이름, 열이름]

print(df.iloc[0,1])
print(df.iloc[1,2])
print(df.loc["짱구","python"])
print()


#여러행, 여러 열 선택
print(df.iloc[0,[1,3]]) #행 열의 순서대로 값이 1개인상인것은 리스트로 묶어서 전달
print(df.iloc[2,[2,1]]) #시리즈 형태
print(df.loc["짱구",["웹표준","java"]])
print(df.loc[["철수","유리"],"python"])
print()

#슬라이싱 사용
print(df.loc['짱구':'철수','python':'C언어']) #데이터 프레임 형태로출력
#2개의 행 3개의열
print(df.iloc[1:,2:,])
print(df.iloc[1:3,::3]) #인덱스 번호 사용하여 원소 접근 시 스텝값 지정가능
print()

#열로만 슬라이싱(행전체)
print(df.loc[:,"python":"C언어"])
print(df.iloc[:,1:3])
```

```python
#산점도 그래프 그리기
#scatter():산점도 그래프
#데이터들간의 상관관계를 파악할때 사용함
#판다스에서 상관관계를 확인하는 .corr()메소드와 함께 사용하며
#수치들간의 상관관계를 확인하는 것이기때문에 숫자형 데이터만 사용가능!!

#df.corr() : 모든 열의 상관계수 확인
#df.['열이름'].corr(): 특정열의 상관관계 확인

plt.figure(figsize=(8,6))

x=['봄','여름','가을','겨울']
y1=[100,200,300,400]
y2=[150,300,200,50]

plt.scatter(x,y1)
plt.scatter(x,y2, s=100, c='red') 

plt.title('계절별 에어컨 가동횟수')

plt.show()
```

```python
#seaborn 데이터를 이용하여 붓꽃 데이터를 활용한 산점도 그래프 만들기
#seaborn 데이터를 제공하는 패키지

import seaborn as sns

iris=sns.load_dataset('iris') 

iris #붓꽃 데이터로드

import matplotlib.pyplot as plt

#대상컬럼
sl=iris['sepal_length'] #꽃받침의 길이
pl=iris['petal_length'] #꽃잎의길이 
#지금 변수안에각각의 열만 떼온 값들을 집어넣어놓은것.
#꽃받침의 길이와 꽃잎의 길이의 상관관계를 알아볼것.

plt.rcParams['font.family']="Malgun Gothic"
plt.figure(figsize=(8,6))
plt.scatter(sl,pl, c='skyblue', s=30, marker='o')

plt.xlabel('꽃받침 길이')
plt.ylabel('꽃잎의 길이')
plt.title('꽃받침의 길이와 꽃잎의길이의 상관관계')

plt.show()

iris[['sepal_length','petal_length']].corr()#데이터프레임에서 열2개선택한다음에
#상관관계 확인할 수 있는 함수사용해서 수치적으로 함수 본것
```

```python
#<상관관계 실습> scatter 산점도그래프 만들기
#sepal_width 와 petal_width의 상관관계 알아보기


iris
sl=iris['sepal_width'] #꽃받침의 두께를 변수에 넣음
pl=iris['petal_width'] #꽃잎의 두께를 변수에 넣음

plt.figure(figsize=(8,6)) #사이즈 설정
plt.rcParams['font.family']="Malgun Gothic" #한글깨짐 방지

plt.scatter(sl,pl, s= 30, c='red') #산점도 그래프 설정

plt.xlabel('꽃받침의 두께')
plt.ylabel('꽃잎의 두께')
plt.title('꽃받침과 꽃잎의 두께간의 상관관계')

plt.show()

iris[['sepal_width', 'petal_width']].corr()
```

```python
import pandas as pd
from palmerpenguins import load_penguins

# 데이터셋 불러오기
penguin = load_penguins()

# 데이터셋의 첫 몇 행 출력
print(penguin.head())
```

```python
#팽귄의 부리의 너비와 길이의 상관관계(서식지별로)
#서식지 별로 나눈다는 조건이 붙었기때문에 서식지안에 어떤 값이 있는지 범주값을 확인해야함

#서식지의 값 확인하기
# print(penguin['island'].value_counts()) 
marker={"Biscoe":"o","Dream":"+","Torgersen":"s"} #마커라는 변수안에 키값으로 서식지를, 벨류로 마커의 종류를넣음
color={'Biscoe':'red','Dream':'blue','Torgersen':'green'} #컬러라는 변수안에 키값으로 서식지, 벨류로 색상넣음

plt.figure(figsize=(20,10))
groups=penguin.groupby('island')
# print(groups) #그룹이라는 객체만 나옴

for i,j in groups:
    plt.scatter(j['bill_length_mm'],j['bill_depth_mm'], s=20, color=color[i], marker=marker[i],label=i)


plt.legend(fontsize=20)
plt.xlabel('부리너비')
plt.ylabel('부리 길이')
plt.title('팽귄의 부리너비와 길이의 상관관계',fontsize=20)

plt.show()

# penguin
```

```python
#<팽귄 종류별로 너비와부리의 상관관계 반복문으로 구하기>

#퍙귄 종류의 범주값 확인
penguin['species'].value_counts()

#그릴 캠퍼스 만들기
plt.figure(figsize=(8,6))

#반복문에서 사용할 마커와 색상 스타일 지정
m_style={'Adelie':'*','Gentoo':'o','Chinstrap':'s'}
c_type={'Adelie':'red','Gentoo':'black','Chinstrap':'green'}

#팽귄 종류대로 그룹화
groups=penguin.groupby('species')

for i,j in groups:
    plt.scatter(j['flipper_length_mm'],j['bill_depth_mm'], s=20, color=c_type[i],marker=m_style[i],label=i)

plt.legend()
plt.title('팽귄종류에 따른 날개길이와 성별의 상관관계', fontsize=20)
plt.xlabel('날개길이')
plt.ylabel('성별')

penguin[['flipper_length_mm','bill_depth_mm']].corr()
```

```python
#원형그래프 그리기

import matplotlib.pyplot as plt
import pandas as pd


figure=plt.figure() #figure라는 변수안에 크기 자동지정해둔것
#원형그래프의 경우 딱히 캔버스대로 나오는것이 아니기때문에 굳이 값을지정해주지않고
#형식상의 캔버스만 만들어둔것

plt.rcParams["font.family"]="Malgun Gothic"


#원형그래프는 다른 그래프와 다르게 비율을 나타내는 그래프이기때문에 
#x,y값이 아니라 비율과 라벨이 들어감. 지금 그걸 따로 변수안에 담아 지정해준것.
#x,y값 리스트로 따로 지정하듯이!



labels=['봄','여름','가을','겨울']
ratio=[50,30,12,28] #원형그래프에 들어가는 비율은 이 숫자자체가 비율이 아니라
#전체 숫자(100)서 각각의 숫자가 어느정도의 비율을 차지하는것인지를 보여줌 


plt.pie(ratio,labels=labels,autopct="%.2f%%",wedgeprops={'linewidth':2,'edgecolor':'black'})
#autopct: 부채꼴안에 표시될 숫자의 형태 지정
#wedgeprops : 부채꼴의 형식지정. 옵션에서 딕셔너리 형태로 지정함
#linewidth:선두께
#edgecolor:선의 가장자리

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("./day09/국내여행_횟수.csv",index_col=[0,1])
df=df.loc['연령']


fig=plt.figure(figsize=(12,16))

num=1

for i in df.columns: #컬럼만 나오도록 선택함 그럼 1월~12월까지 나오는데
    df.months=df[i]
    # print(df.months #그값을 이용해서 컬럼의 열선택으로 사용함. 열선택하면 시리즈 형태로 열이름이 name으로 들어가고 행값이 나옴
    axes1=fig.add_subplot(4,3,num)
    axes1.pie(df.months,labels=df.months.index,autopct="%2.f%%",startangle=240)
    axes1.set_title(f"{i}의 연령대별 여행비율", fontsize=13)
    axes1.legend(loc="upper right", fontsize=7)
    num+=1


plt.show()
```

```python
#<반복문 원형데이터 실습>

import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("./day09/국내여행_횟수.csv",index_col=[0,1])
df=df.loc['연령']
df

fig=plt.figure(figsize=(12,16))

num=1

for i in df.columns:
    df.months=df[i]
    axes1=fig.add_subplot(4,3,num)
    axes1.pie(df.months,labels=df.index,autopct="%.2f%%",startangle=90)
    axes1.set_title(f"{i}의 연령별 여행횟수")
    axes1.legend(loc="upper right")
    num+=1

plt.show()

plt.savefig("./day09/월별_여행비율")
```

```python
#regplot : 회귀선이 있는 산점도 그래프
#회귀분석 : 서로 영향을 주는 2개의 변수 관계를 파악해서 종속변수의 값을 예측하는것
#fit_reg=False로 설정하면 회귀선 보지 않을 수 있음

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

titanic=sns.load_dataset('titanic')

titanic

fig=plt.figure(figsize=(15,5))
axes1=fig.add_subplot(1,2,1)
axes2=fig.add_subplot(1,2,2)

sns.set_style('darkgrid')
sns.regplot(data=titanic,  #어떤 데이터를 쓸것인지
           x='age',
           y='fare',
           ax=axes1) #위치설정

sns.regplot(data=titanic,
         x='age',
         y='fare',
         ax=axes2,
         fit_reg=False,
        color='red')
```

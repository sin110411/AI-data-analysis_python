```python
import pandas as pd
#데이터 프레임 생성
stu_df=pd.DataFrame({'수학':[80,75,90],'영어':[90,60,70],'국어':[70,90,85]},
               index=[f"학생{x}"for x in range(1,4)])

stu_df


#과목마다 평균점수 (열마다 평균점수 확인)
print(stu_df.max())
print()

#가장 높은 국어점수
print(f"가장 높은 국어점수:  {stu_df['국어'].max()}\n")



#가장 낮은 수학점수
print(f"가장 낮은 수학점수: {stu_df['수학'].min()}")


#여러 열선택
round(stu_df[['국어','영어']].mean(),2)
print()

#각 과목의 표준편차
print(stu_df.std())
print(f"수학의 표준편차: {stu_df['수학'].std()}" )
```

```python
#<실습>


df=pd.read_csv("./day09/국내여행_횟수.csv",header=0, index_col=[0,1])
df=df.loc['연령']

#1) 20대의 월별 최대 여행 횟수
print(f"20대 월별 최대 여행 횟수: {df.loc['20대'].max()}" )
print()

#2) 월별 총 여행 횟수 출력
print(df.sum())
print()

#3) 연령대별 여행횟수 최대여행횟수
df_t=df.transpose()
print(df_t.max())
print()

#4) 연령대별 여행횟수 평균값
print(round(df_t.mean(),2))
print(df.transpose().mean().round(2).astype(int))

#따로 변수 안에 넣지 않으면 이렇게 라운드도 괄호안에 일반 숫자로 넣을 수 있고, 여기서 정수로만 뽑고싶다면 
#astype을 이용하여 int형으로 만드는것도 가능

print(df.transpose().mean().round(3).astype(int))
```

```python
#웹크롤링

import requests #요청 패키지
from bs4 import BeautifulSoup as bs #HTML코드에서 필요한 정보만  추출하는 패키지

url="https://news.naver.com/" #url주소 확인
response=requests.get(url) #get을 사용해서 응답을 요청. 그에 대한 응답을 response에 저장
print(response.status_code) #상태코드로 웹크롤링이 가능한 상태인지 확인
#200이 출력되었으므로 웹크롤링을 할 수 잇는 상태
print(response.text)

html=response.text #응답받은 HTML코드를 추출할 수 있도록 변환
#코드의 트리구조 파악
soup=bs(html,"html.parser")#코드의 트리구조 파악
print(soup)
```

```python
#find('태그명') : 해당 태그명 중에서 첫번째꺼 가져오는것


span=soup.find("span") #soup안에 트리구조를 파악한 코드가 들어있음. 여기서 span으로 시작하는 태그중
#첫번째 값 가지고 온것
print(span) #그걸 출력한것

print(span.text) #네이버 출력
#span태그 안에서 첫번째값을 가지고 오는것 바로 정보빼오기 가능
#find는 맨 첫번째 나오는 태그명 하나만 추출하기 때문에 바로 내용을 빼올 수 있음
print()
```

```python
#객체.find_all(태그명) : 해당 태그명의 값 한번에 리스트형태로 다 빼오기

span1=soup.find_all("span")
print(span1) # soup안에 트리구조로 되어있는 구문에서 span으로 시작하는 모든 태그들을 다가져옴
#자세히보면 <span>내용</span>, <span>,...이런 구조로 되어있음
#그렇기 때문에 내용을 바로 뺄 수 없음. 고로 인덱스 번호를 사용하여 값을 빼내야함
print(span1[0])  #가장 첫번째를 가지고 오라고했기때문에 아직 내용 빼오기 전인 print(span)과 같은 결과
print(span1[2])  #뉴스가 있는 태그를 가지고온것
print()
print(soup.find_all("span")[2].text)
print()

#언론사별 이라는 문자만 추출하기
#클래스명 이용하기
str1=soup.find_all("span",class_="Nitem_link_menu")[0]
print(str1.text)
```

```python
#네이버 스포츠 페이지에서 야구 순위표 추출하기

import requests
from bs4 import BeautifulSoup as bs

#주소 입력하기
url="https://sports.news.naver.com/kbaseball/record/index?category=kbo"

#주소 상태 확인하기
response=requests.get(url)
print(response.status_code) #200이 출력되었으므로 웹크롤링 가능

#html로 변환하기
html=response.text #그냥 response를 넣는게 아니라 text(우리가 읽을 수 잇는 형태로 바꿔주기)
soup=bs(html,"html.parser") #추출할 수 있는 형태로 바꿔주기

#내용 추출하기
#find 사용

#탐순위라는 문자 추출하기
data1=soup.find("h4")
print(data1.text)


#find_all사용하기
data2=soup.find_all("strong")
print(data2[4].text)
print()


#순위표 내용 가져오기
ranking_table=soup.find("tbody",id="regularTeamRecordList_table")
#tbody태그안에 id속성값이 한개만 있는것을 확인후에 변수 대입
print(ranking_table)
print()

#표 행가져오기
rows=ranking_table.find_all("tr")
# print(rows.text) all로 가져왔기때문에 값이 여러개여서 text를 사용할 수 없음.
#이럴경우 하나씩 뽑거나 해야하는데 이게 번거롭기 때문에 반복문을 사용해서 
#값을 순서대로 하나씩 뽑아줄것임
print()

#반복문을 사용하여 행 추출
for i in rows:
    th=i.find("th").text.strip()
    span=i.find_all("span")[1].text.strip() #구단 명
    print(f"{th}위 {span}")
```

```python

#크롤링한 내용을 text파일로 저장하기

ranking_table=soup.find("tbody",id="regularTeamRecordList_table")

#파일열기
with open("./ranking.txt","w",encoding='utf-8') as f:
    rows=ranking_table.find_all("tr")

    for i in rows:
        th=i.find("th").text.strip() #팀순위 #strip(): 양쪽의 불필요한 공백 제거 메소드.  문자열메소드에서 아무것도 안쓰면 공백제거 됨
        span=i.find_all("span")[1].text.strip() #구단명 빼오기 
        print(f"{th}위 {span}")
        f.write(f"{th}위 {span}\n") #줄바꿈 안되서 따로 넣어줌

print("순위와 구단을 파일에 저장했습니다")
```

```python
#<실습>

import requests
from bs4 import BeautifulSoup as bs

#1)네이버 기사 웹크롤링 후 
#주소 확인
url="https://n.news.naver.com/article/469/0000817526?cds=news_media_pc&type=editn"

#주소 상태코드 확인
response=requests.get(url)
# print(response.status_code) #코드 200확인


#html코드로 변환
html=response.text

#트리구조 파악
soup=bs(html,"html.parser")
# print(soup)

#2) 기사 제목 
title=soup.find("h2").text
print(title)


#3)기사 내용 따로 가져와서 호출
content=soup.find("article").text
print(content)


#4)메모장에  .txt파일로 저장하기
with open("./아귀_기사.txt","w",encoding='utf-8') as f:
    f.write(f"기사제목: {title}\n")
    f.write(f"기사본문: {content}\n")


print("기사 파일이 정상적으로 저장되었습니다")
```

```python
#실습_기사내용 하나 더 추출하기

#패키지 호출
import requests
from bs4 import BeautifulSoup as bs

#url주소 확인
url="https://n.news.naver.com/article/666/0000049358?cds=news_media_pc&type=editn"
response=requests.get(url)
# print(response.status_code)  #상태코드 200확인

#html코드 변환
html=response.text
# print(html) #변환성공. 변환할때 text메소드를이용해서 우리눈에 보기 쉡게 바꿈

#bs패키지 활용하여 코드의 트리구조 분석하고 필요한 자료만 빼오기
soup=bs(html,"html.parser") #퍼서=파싱
# print(soup)

#find구문 이용하여 기사 제목 추출
a_title=soup.find("h2").text
print(a_title)

#기사 본문 추출 근데 class를 활용한..!
a_content=soup.find(class_="go_trans _article_content").text
# print(a_content)

#웹크롤링한 기사제목과 기사내용 txt파일에 저장하기
with open("./청년빈곤심각.txt","w",encoding='utf-8') as f:
    f.write(a_title)
    f.write(f"[기사내용]\n")
    f.write(f"{a_content}")
```

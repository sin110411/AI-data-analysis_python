```python
#데이터 프레임연결 : pd.concat(), merge(), join()
#1) pd.concat(데이터 프레임 리스트) : 리스트 요소로 있는 데이터 프레임 연결
#옵션

import pandas as pd

#데이터 프레임 생성
#리스트 컴프리헨션으로 0~3까지의 리스트 생성
df1=pd.DataFrame({'a':[f"a{x}" for x in range(4)],
                 'b':[f"b{x}" for x in range(4)],
                 'c':[f"c{x}" for x in range(4)]},
                 index=['zero','one','two','three'])

print(df1)

df2=pd.DataFrame({'b':[f'b{x}'for x in range(2,6)],
                 'c':[f"c{x}"for x in range(2,6)],
                 'd':[f"d{x}"for x in range(2,6)]},
                 index=['two','three','four','five'])

print(df2,'\n')

#2개의 데이터프레임을 행방향으로 이어붙여 연결하기
result1=pd.concat([df1,df2]) #리스르 요소로 데이터프레임2개가 전달됨
print(result1)
print()
#df1아래에 df2가 연결됨
#df1에는 d열 데이터가 없으므로 연결될때 d열값이 NaN으로 채워짐
#각자 서로에게 없는 값은 NaN으로 채워짐

#ignore_index=True 옵션: 기존 인덱스를 무시하고 새로운 인덱스로 설정, 정수형 인덱스(0~)
result2=pd.concat([df1,df2],ignore_index=True)
print(result2) #기존의 행인덱스 무시되고 숫자 0부터 7꺄지 번호로된 인덱스가 들어감
print()

#열기준으로 데이터 프레임 연결
result3=pd.concat([df1,df2],axis=1)
print(result3)
print()
#zero, one은 df1에만 존재하고 four, five는 df2행에만 존재
#어느 한쪽에만 데이터가 존재하는 경우 : 비어있는 위치의 값은NaN으로 설정된다

#join='inner' :연결할 데이터 프레임의 교집합기준으로 구성
result4=pd.concat([df1,df2],join='inner') #디폴트 값으로 행기준이 설정되어있어서 세로로 출력됨
print(result4)#행기준으로 연결할때 공통되는 열인 b,c기준으로 연결
print()

#join='inner' : 열기준으로 교집합하기
result5=pd.concat([df1,df2],join='inner',axis=1)
print(result5) #열방향으로 설정했을때는 공통되는 행인 two, three를 기준으로 출력!
print()

#join='outer': 열이나 행 레이블의 합집합으로 데이터 프레임 구성. 데이터 프레임 연결의 디폴트 값!
result6=pd.concat([df1,df2],join='outer')
print(result6) #result1과 똑같은 결과를 반환. join='outer'는 디폴트로 들어가있기떄문!
```

```python
#데이터 프레임 병합
import pandas as pd
#데이터 프레임 병합 : pd.merge(df_list,df_right,how='any',on=None)

df_left=pd.DataFrame({'번호':[1,2,3,4],
                     '이름':['짱구','철수','유리','맹구'],
                     '나이':[5,6,7,6]})


df_right=pd.DataFrame({'번호':[1,3,5,6],
                  '학생이름':['짱구','유리','둘리','또치'],
                  '점수':[60,85,70,90]})

print(df_left,'\n')
print(df_right,'\n')

result1=pd.merge(df_left,df_right)
print(result1) #how=inner : 기준이 되는 열의 데이터가 양쪽 프레임에 공통으로 존재하는 경우에만 추출
#기준이 되는 열(번호)의 데이터가 양쪽 데이터 프레임에 똑같이 존재하는 경우 (1,3)
print()

#데이터 프레임 병합(how='outer':key열 기준 합집합)
result2=pd.merge(df_left,df_right,how='outer',on='번호')
print(result2)
#번호기준으로 데이터 프레임을 합집합으로 병합
#왼쪽데이터 프레임에 5,6번 없음 =>해당번호의 이름, 나이, 열데이터 NaN
#왼쪽데이터 프레임에 2,4번 없음 =>해당번호의 학생이름, 점수 데이터 NaN
print()

#데이터프레임 병합 : 왼쪽데이터 기준으로 병합
result3=pd.merge(df_left,df_right,how='left')
print(result3,'\n') #왼쪽데이터프레임의 번호열 기준으로 병합


#데이터프레임 병합: 오른쪽데이터 기준으로 병합
result4=pd.merge(df_left,df_right,how='right')
print(result4) #오른쪽 기준으로 데이터 병합
```

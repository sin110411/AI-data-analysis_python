```python
#시계열 데이터 활용예시1
#그래서 이렇게 만든 시계열 데이터를 뭐 어떻게 활용하느냐??

import pandas as pd

df=pd.read_csv('./online_retail.csv')
df.head()

#문자열 데이터 -> Timestamp 객체로 바꾸기
df['new_date']=pd.to_datetime(df['InvoiceDate'])
df

#기존의 문자열 데이터 열 삭제하기
df.drop('InvoiceDate', axis=1,inplace=True)
df.head()

#시계열 데이터의 dt속성을 사용하여 연,월,일 따로구분
df1=df.copy()
df1['Year']=df1['new_date'].dt.year
df1['Month']=df1['new_date'].dt.month
df1['Day']=df1['new_date'].dt.day
df1
```

```python
#시계열 데이터 활용2
#Timestamp를 period로 변환하여 연,월,일 분리

df2=df.copy()
df2['Date_Year']=df['new_date'].dt.to_period(freq="A")
df2["Date_day"]=df['new_date'].dt.to_period(freq="D")
#period는 결과가 Timestamp와 똑같아 보이지만 '동안'의 개념을 가지고있음 2010년 12월 01일 하루 동안!

#dt.to_period속성으로 빼온 값을 인덱스로 지정하기
df2.set_index('Date_day',inplace=True)
#원래 new_date에 만들어준 날짜 시계열 데이터를 일자만 빼와서 인덱스로 설정하면 시계열데이터를 인덱스로 가진 데이터프레임을 만들 수 있음
df2


#시계열 데이터 활용2
#날짜사이의 간격을 계산할 수 있다(시점끼리 계산가능)
open_day=pd.to_datetime("2018-12-01")
print(open_day,type(open_day)) #pd.to_datetime을 이용해서 문자데이터인 '2018-12-01'을 타임스탬프로 바꿔 변수안에 저장

#데이터 프레임에있는 period 인덱스를 datetime 형식으로 변환
df2.index = df2.index.to_timestamp()

#time_laps
df2['time_lapse']=df2.index-open_day
print(df2)

df2
```

```python
#시리즈의 함수 매핑
#sr.apply함수 사용
#같은 크기의 시리즈를 반환

import pandas as pd
import seaborn as sns

df=sns.load_dataset('titanic')
df

#사용자 정의함수 복습하기
def add_ten(element):
    return element+10

#시리즈 객체에 사용자 정의함수 매핑하기 : sr.apply
df=df[['age','fare']]

#사용자 정의함수 호출하기
sr1=df['age'].apply(add_ten)
#원래 사용자 정의함수를 사용할때 매개변수가있다면 함수명뒤에 소괄호안에 인수를 전달해주어야함
#근데 sr.apply사용할때는 이미 객체를 전달해주고 있음. 따로 소괄호를 이용해서 인수를 전달해주지 않아도 됨

print(sr1,type(sr1))

#사용자 정의함수2
def add_two(x,y):
    return x+y


sr2=df['age'].apply(add_two,y=5)
sr2
#이 경우는 전달해주어야하는 인수가 2가지. 
#이런경우에는 객체의 원소말고도 인수가 한개 더필요하기떄문에 함수,매개변수=값의 형태로 
#인수를 전달해주어야 한다


#람다식활용(일시적인 값 확인을 위해서는 람다를 많이 사용-람다는 익명함수(일회용함수이기때문에!))
#람다 : (lambda 매개변수 : 리턴값)(인수,인수)
sr3=df['age'].apply(lambda x : x +10) #람다식에서의 인수는 df['age']의 원소
print(sr3)
```

```python
#데이터 프레임 함수 매핑
#(1) 데이터 프레임 원소에 함수 매핑
import pandas as pd
import seaborn as sns

df=sns.load_dataset('titanic')
df=df[['age','fare']]
df

#데이터 프레임 원소에 함수 매핑
#df.map(함수)
df_map=df.map(add_ten)
df_map=df.map(lambda x : x+10) #값은 사용자 함수와 람다와 똑같이 적용
print(df_map,type(df_map)) #데이터 프레임으로 값 반환

df_map1=df.map(lambda x : x**2)
df_map1
#데이터 원소 하나하나에 전부 함수 매핑이 된것
```

```python
#데이터 프레임 객체에 함수 매핑 : 데이터 프레임, 시리즈, 값(스칼라)의 형태로 값 반환
#df.pipe(함수)

#사용자 정의함수
#각열의 NaN값 찾기 : 데이터 프레임을 전달하면 데이터 프레임을 반환
def null_value(df):
    return df.isnull()

print("====데이터 프레임으로 반환====")
result_df=df.pipe(null_value)
print(result_df,type(result_df))
```

```python
#시리즈(sr)type(df.isnull().sum()) 타입시리즈로 반환
def null_count(sr):
    return null_value(sr).sum() #null_value에서 데이터프레임으로 값이 나온걸활용해서
    #갯수를 구하는 사용자 정의함수로 다시 만들어준것. 사용자 정의함수에 안에 또 다른 사용자 정의함수를
    #넣을 수 있음

result_sr=df.pipe(null_count)

print("====시리즈로 반환====")
print(result_sr,type(result_sr)) #시리즈로 반환
#함수에서 리턴값에 어떤걸 넣느냐에 따라서 데이터 프레임이 나올 수 도 있고 시리즈가 반환될 수도 있음
```

```python
#값으로 반환(스칼라로 반환)

#age와 fare의 누락데이터 갯수를 null_count를 가지고 그 두개를 더해서 최종적인 누락데이터의 갯수를 구하는 함수 정의
def null_total(element):
    return null_count(element).sum()

result_cnt=df.pipe(null_total)
print("====스칼라 데이터로 반환====")
print(result_cnt,type(result_cnt)) #스칼라로 반환
```

```python
#<실습>

import pandas as pd
import seaborn as sns

df=sns.load_dataset("mpg")

df
#1) cylinders(실린더)별 배기량(displacement)을 나타내는 per_cylinders열 만들기
#-행마다 함수 적용하기
df['per_cylinders']=df.apply(lambda row: row['cylinders']/row['displacement'],axis=1)
df

#1번 함수적용해서 풀기
def result(df):
    return df['cylinders']/df['displacement']

df['per_cylinders1']=df.apply(result,axis=1)
print(df)

#2) weight열에서 데이터가 짝수면 '짝수', 홀수면 '홀수' 출력
df['홀짝']=df['weight'].apply(lambda x: "짝수" if x%2==0 else "홀수")
df[['홀짝','weight']]

#3)mpg데이터에서 acceleration열과 model_year열만 따로 분리하여 데이터프레임을 
#만든 후 새로만든 데이터 프레임 전체 데이터에 10씩 더한 결과 출력 

df1=df[['acceleration','model_year']]

df2=df1.pipe(lambda x: x+10)
print(df2,type(df2))

df1
```

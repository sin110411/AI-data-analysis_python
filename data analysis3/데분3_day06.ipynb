```python
#데이터 프레임 결합
#객체.join :  두 데이터 프레임을 '행기준'으로 결합
#객체1.join(객체2, how='left or right')

import pandas as pd

#데이터 프레임 결합
#1) 데이터프레임 생성

df_left=pd.DataFrame({'번호':[1,2,3,4],
                     '이름':['짱구','철수','유리','훈이'],
                     '나이':[5,6,7,6]})

df_left.set_index('번호',inplace=True)
print(df_left,'\n')

df_right=pd.DataFrame({'번호':[1,3,5,6],
                      '학생이름':['짱구','유리','둘리','또치'],
                      '점수':[60,85,70,90]})

df_right.set_index('번호',inplace=True)
print(df_right,'\n')

#데이터 프레임 결합
result=df_left.join(df_right) #join은 행기준. 번호 행을 기준으로 결합
print(result) #df_left의 번호 행을 기준으로 결합.
#df_left에 학생이름과 점수가 없음. join되면서 열이 합쳐졌는데 5,6번은 df_left기준 조인되는 값이 없기때문에
#아무것도 출력되지 않고 NaN값 출력

#join의 합집합, 교집합
#outer: 합집합, inner: 교집합(join뿐만아니라 모든 결합에서 사용됨)
result1=df_left.join(df_right,how='outer')
print(result1,'\n')

result2=df_left.join(df_right,how='inner')
print(result2,'\n') #교집합. 완전히 겹치는 데이터만 출력

#결합되는 기준 df 설정
#원래는 객체1의 자리에 기준을 원하는 df를 작성하면 따로 how에 옵션설정하지 않아도 자동으로 객체1의 df를 기주능로
#조인이 적용되지만 객체1에 적었더라도 how를 통해 기준df를 정할 수 있음
result3=df_left.join(df_right,how='left')
print(result3,'\n')

result4=df_left.join(df_right,how='right')
print(result4)
```

```python
#그룹연산메소드(적용 - 결합단계)
import pandas as pd
import seaborn as sns

df=sns.load_dataset('titanic')
df=df.loc[:,['age','sex','class','fare','survived']]
df


#class열 기준으로 그룹화하기(그룹객체만들기 --> 1단계 )
#1) class열의 고유값 확인하기
print(df['class'].value_counts(),'\n')

#2)class열 기준 그룹화만들기
groups=df.groupby('class')
# print(groups) #그룹객체반환

#3)반복문으로 그룹된 값 빼주기 (그룹객체만들기 완료)
# for key,group in groups:
#     print(key)
#     print(group)
# print()

#4)통계함수로 집계
#4-1)각그룹에 대한 age,fare의 표준편차를 집계하여 데이터프레임으로 변환
std_all=groups[['age','fare']].std() #3번에서 그룹객체만들어준 후에 groups 변수에 담음
#group변수안에는 3가지의 그룹객체가 있는것. 거기서 groups[['age','fare']]의 의미는 각 그룹마다있는
#열 두개를 선택해서 각각 그룹별로 두 행의 표준편차를 보겠다고 한것!
#여기서 열두개선택했기떄문에
print(std_all,type(std_all),'\n')#결과는 데이터프레임으로 출력


#4-1)#각 그룹에 fare열의 표준 편차 집계
std_fare=groups['fare'].std()
print(std_fare,type(std_fare),'\n')#열을 1개만 선택했기때문에 시리즈로 출력
```

```python
#사용자 정의함수 그룹객체에 적용: 그룹객체.agg()

#1)사용자 정의함수 그룹객체에 적용 : 그룹객체.agg()
def max_min(group):#매개변수의 인수로 그룹이 들어올것이라는것을 알려줌
    return group.max()-group.min()  #그룹마다 큰값에서 작은값 빼기

#그룹별 사용자 정의함수 지정
agg_maxmin=groups[['age','fare']].agg(max_min) #열 두개 선택했으니 데이터프레임
print(agg_maxmin)

#위와같은 식 람다식활용
print((lambda group : group.agg(max_min)))
agg_maxmin=groups[['age','fare']].agg(lambda group: group.max()-group.min())

agg_maxmin

#전체 그룹에서 get_group(키) 이용하여 해당 그룹의 원하는 내용만 가져오기
group1=groups.get_group('First')
print()
print(group1)
print(group1[['age','fare']].max(),'\n')
print(group1[['age','fare']].min(),'\n')
print(group1[['age','fare']].max()-group1[['age','fare']].min(),'\n')
```

```python
#여러함수를 각 열에 동일하게 적용하여 집계하기
#그룹객체.agg([함수1, 함수2,...])

print("===같은 열에 여러함수 적용===")
agg_all=groups[['age','fare']].agg(['min','max'])
print(agg_all,'\n')
agg_all1=groups[['age','fare']].agg(['sum','median'])
print(agg_all1,'\n')

#각 열마다 다른 함수를 적용하여 집계
#fare열에는 2개의 함수(min,max), age열에는 mean함수 적용
print("===각 열마다 다른 함수를 적용===")
agg_sep=groups[['age','fare']].agg({'age':'mean', 'fare':['min','max']})
print(agg_sep,'\n') #age는 평균만 따로나오고 요금은 최소최대 둘다 나옴!
agg_sep1=groups[['age','fare']].agg({'age':'std','fare':['std','median']})
print(agg_sep1)
```

```python
#<과제>
import seaborn as sns

#seaborn라이브러리에서 diamonds데이터 불러오기
df=sns.load_dataset('diamonds')
df

# 다이아몬드 데이터 프레임에서 cut품질별로 그룹화한 후 그룹별로 가격의 평균을 계산
print(df['cut'].value_counts())
groups=df.groupby('cut')
print()
print("======cut품질별 가격의 평균=========")
print(groups['price'].mean(),'\n')

# 컷팅등급이  ‘Good’,인 그룹의 다이아몬드의 무게 합계 계산
cut_g=groups.get_group('Good')
cut_g=cut_g['carat']
print("====컷팅등급이  ‘Good’,인 그룹의 다이아몬드의 무게 합계=====")
print(cut_g.sum(),'\n')

# cut별로 그룹화 후 그룹마다 최대가격과 중간값의 차이를 구하고
d_maxmedian=groups['price'].agg(lambda x : x.max()-x.median())
print("====cut별로 그룹화 후 그룹마다 최대가격과 중간값의 차이=====")
print(d_maxmedian,'\n')
# 그룹마다의 다이아몬드 무게의 최소값과 평균 구하기
print("====그룹 별 다이아몬드 무게의 최소값과 평균=====")
d_minmean=groups['price'].agg(['min','mean'])

# 다이아몬드 데이터를 컷팅 등급과 색깔로 그룹화한 후 그룹마다 가격의 중간값이 3000달러 보다 많은 그룹들만 출력
groups1=df.groupby(['cut','color'])

fi_agg=groups1.filter(lambda x : x['price'].median()>=3000)
print("====cut, color로 그룹화 후 가격의 중간값이 3000달러 이상인 그룹=====")
print(fi_agg[['cut','color','price']].value_counts())
```
